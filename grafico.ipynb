```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_diabetes
```

```python
#cargar el data set
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target

#mostrar primeras filas
df.head()
```

```python
#distribucion de la variBLE objetivo (target)

plt.figure(figsize=[6, 3])

#tipo de grafico que queremos , en este caso un histograma con histplot
sns.histplot(df['target'], bins=20, kde=True)
plt.title('Distribucion de la variable objetivo')
plt.xlabel("progresion de la diabetes")
plt.ylabel("frecuencia")
plt.show()
```

```python
plt.figure(figsize=(6,3))
df_melt = df.melt(id_vars="target", var_name="variable", value_name="valor")
sns.boxplot(x="variable", y="valor", data=df_melt)
plt.title("distribucion de la variable independientes")
plt.show()
```

```python
#grafico boxplot
plt.figure(figsize=[6, 3])

#tipo de grafico que queremos , en este caso un boxplot con boxplot
sns.boxplot(x=pd.qcut(df['bmi'], 4), y="target", data=df)
plt.title('progresion de la deabetes segun los datos de bmi (boxplot)')
plt.xlabel("cuartiles de bmi")
plt.ylabel("progresion de la diabetes")
plt.show
```

```python
sns.scatterplot(x="bmi", y="target", data=df, alpha=0.06)
plt.title("relacion entre bmi y progresion de la diabetes")
plt.xlabel("BMI(indice de masa corporal)")
plt.ylabel("progresion de la diabetes")
plt.show()
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x="bmi", y="target", data=df, alpha=0.06)
plt.title("relacion entre bmi y progresion de la diabetes")
plt.xlabel("BMI(indice de masa corporal)")
plt.ylabel("progresion de la diabetes")
plt.show()
```

```python
plt.figure(figsize=[6, 3])

#tipo de grafico que queremos , en este cao un histograma con histplot
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('matriz de correlacion')
plt.show()
```

```python
#tomamos solo algunas variables para no saturar el grafico
subset_df = df[["bmi", "bp", "s5", "target"]]

sns.pairplot(subset_df, diag_kind= "kde")
plt.suptitle('relaciones multiples entre variables', y=1.02)
plt.show()
```

```python
#grafico violinplot
plt.figure(figsize=(8,5))
sns.violinplot(x=pd.qcut(df["bmi"], 4), y="target", data=df)
plt.title("progresion de la diabetes según datos de bmi (violinplot)" )
plt.xlabel("cuartiles de bmi")
plt.ylabel("progresion de la diabetes")
plt.show()
```

```python
#comparacion en un grafico mas visual
plt.figure(figsize=(7,5))
sns.kdeplot(df["bmi"], shade=True, label="bmi")
sns.kdeplot(df["bp"], shade=True, label="bp")
plt.title("comparacion de distribuciones [KDE]")
plt.xlabel("valor normalizado")
plt.ylabel("densidad")
plt.legend()
plt.show()
```
